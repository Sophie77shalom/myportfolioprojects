```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file_path = "/kaggle/input/marvel-wikia-dataset/marvel-wikia-data.csv"
my_data = pd.read_csv(file_path)
my_data
```

Line chart showing evolution in comic books.


```python
#line chart
sns.lineplot(data=my_data)
```

```python
plt.figure(figsize=(16, 8))
plt.title("Comic Books")
sns.lineplot(data=my_data)
```

```python
list(my_data.columns)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Comic Books")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=my_data['EYE'], label="EYE")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=my_data['page_id'], label="page_id")

# Add label for horizontal axis
plt.xlabel("Year")
```

Bar Charts and Heatmaps

```python
# Bar chart
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Comic Books")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=my_data.index, y=my_data['EYE'])

# Add label for vertical axis
plt.ylabel("Year")
```

```python
# Bar chart
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Comic Books")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=my_data.index, y=my_data['page_id'])

# Add label for vertical axis
#plt.ylabel("Year")
```

```python
#heatmap
# Select relevant columns and fill missing values
heatmap_data = my_data.pivot_table(
    index='page_id',
    columns='ALIGN',  # You can also try 'SEX' or other categorical columns
    values='APPEARANCES',
    aggfunc='sum'
).fillna(0)  # Replace NaN with 0

# Plot the heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm")

# Add labels and title
plt.title("Appearances by Alignment and Page ID")
plt.xlabel("Alignment")
plt.ylabel("Page ID")
plt.show()
```

Scatter Plots

```python
#Scatter Plot
sns.scatterplot(x=my_data['Year'], y=my_data['page_id'])
```

```python
#regression line

sns.regplot(x=my_data['Year'], y=my_data['page_id'])
```

```python
#Color-coded scatter plots

sns.scatterplot(x=my_data['Year'], y=my_data['APPEARANCES'], hue=my_data['page_id'])
```

```python
#lmplot
sns.lmplot(x='Year', y='page_id', hue='Year', data=my_data)
```

```python
#swarmplot
sns.swarmplot(x=my_data['Year'],
              y=my_data['APPEARANCES'])
```

Histogram

```python
# Histogram 
sns.histplot(my_data['Year'])
```

kernel density estimate (KDE) plot

```python
#kdeplot
sns.kdeplot(data=my_data['Year'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=my_data['Year'], y=my_data['page_id'], kind="kde")
```
